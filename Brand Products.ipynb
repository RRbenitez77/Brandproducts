```python
r=3
c=3
it=1
for i,j in list(enumerate(df.columns)):
    plt.suptitle("Visualizing the variables")
    plt.subplot(r,c,it)
    if df[j].dtype=='object':
        sns.countplot(x=df[j])
    else:
        sns.kdeplot(df[j])
        plt.grid()
    it+=1
plt.tight_layout()
plt.show()
#set Grid for Subplots and Itterate over Colums and create plots
```

```python
df.head()
```

```python
r=2
c=3
it=1
for i in df.select_dtypes(include='object').columns:
    plt.subplot(r,c,it)
    df.groupby(i)['Rating'].mean().plot(kind='pie',autopct='%0.2f%%')
    it+=1
plt.tight_layout()
plt.show()
```

```python
r=2
c=3
it=1
for i in df.select_dtypes(include='object').columns:
    plt.subplot(r,c,it)
    df.groupby(i)['Price'].mean().plot(kind='pie',autopct='%0.2f%%')
    it+=1
plt.tight_layout()
plt.show()
```
